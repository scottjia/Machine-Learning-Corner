```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
# Set default Seaborn style
sns.set()

# Plot histogram of versicolor petal lengths
iris_data = load_iris()
targets = iris_data.target_names.size
model = KMeans(n_clusters = targets)
model.fit(iris_data.data)
KMeans(algorithm='auto')
labels = model.predict(iris_data.data)

xs = iris_data.data[:,0]
ys = iris_data.data[:,2]

centroids = model.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,2]
centroids_labels = model.predict(centroids)

Label2Color={0: "r", 1:"g", 2:"b"}  
colors = [Label2Color[l] for l in labels]
centroids_colors = [Label2Color[l] for l in centroids_labels]

plt.scatter(xs, ys, c=colors, alpha=0.2)
plt.scatter(centroids_x, centroids_y, c=centroids_colors,marker="D", s=80)

plt.show()
```
