```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip):
    x = np.linspace(0, 14, 100)
    fig,ax=plt.subplots()
    for i in range(1, 7):
        ax.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
    fig.show()        

sinplot(1)
sinplot(3)
```
